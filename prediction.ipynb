```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
import pickle
import numpy as np
```

```python
from tensorflow.keras.models import load_model
import pickle

# Load the pre-trained model
model = load_model('model.h5')

# Load encoder and scaler
with open('Label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender = pickle.load(file)

with open('OneHot_Encoder.pkl', 'rb') as file:
    one_hot_encoder = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
    

```

```python
input_data={
'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}

input_data=pd.DataFrame([input_data])
```

```python
geography_encoded = one_hot_encoder.transform([[input_data['Geography']]])
geography_df = pd.DataFrame(geography_encoded, columns=one_hot_encoder.get_feature_names_out(input_features=['Geography']))
print(geography_df.head())


```

```python

input_data = pd.concat([input_data, geography_df], axis=1)
```

```python
input_data=input_data.drop('Geography',axis=1)
```

```python
label_encoder_gender
```

```python
input_data['Gender'] = label_encoder_gender.transform(input_data['Gender'])
```

```python
scaler
```

```python
## Scaling the input data
input_scaled=scaler.transform(input_data)
```

```python
prediction=model.predict(input_scaled)
```

```python
prediction_proba=prediction[0][0]
prediction_proba
```
